```python
import pandas as pd
```

```python
# CSV path
file_path1 = 'Resources/201909-citibike-tripdata.csv'

# Read the CSV file into a DataFrame
sept_2019_df = pd.read_csv(file_path1)

# Print df
sept_2019_df
```

```python
# CSV path
file_path2 = 'Resources/201910-citibike-tripdata.csv'

# Read the CSV file into a DataFrame
oct_2019_df = pd.read_csv(file_path2)

# Print df
oct_2019_df
```

```python
# CSV path
file_path3 = 'Resources/201911-citibike-tripdata.csv'

# Read the CSV file into a DataFrame
nov_2019_df = pd.read_csv(file_path3)

# Print df
nov_2019_df
```

```python
# Concat all df
citibike_df = pd.concat([sept_2019_df, oct_2019_df, nov_2019_df], ignore_index=True)

# Print
citibike_df
```

```python
citibike_df['gender'].replace({0:'Unknown',1:'Male',2:'Female'}, inplace=True)
citibike_df
```

```python
# Specify path of output
csv_file_path = 'citibike.csv'

# Export the DataFrame to a CSV file
citibike_df.to_csv(csv_file_path, index=False)
```
